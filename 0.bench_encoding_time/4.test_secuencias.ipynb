```python
import random
import pandas as pd

def sequence_gen(nucleotidos, longitud):
    """Genera una secuencia aleatoria de nucleótidos de longitud específica"""
    return ''.join(random.choices(nucleotidos, k=longitud))

def generador_secuencias(n, l, nucleotidos="ACGNT"):
    """Genera n secuencias de longitud l con nucleótidos específicos"""
    temp = []
    for i in range(n):
        temp.append(sequence_gen(nucleotidos=nucleotidos, longitud=l))
    return temp

def generador_secuencias_variables(n, min_l, max_l, nucleotidos="ACGNT"):
    """Genera n secuencias con longitud variable entre min_l y max_l"""
    temp = []
    for i in range(n):
        longitud = random.randint(min_l, max_l)
        temp.append(sequence_gen(nucleotidos=nucleotidos, longitud=longitud))
    return temp

def aplicar_padding(secuencias, longitud_objetivo=1600, caracter_padding='N'):
    """Aplica padding al final de secuencias menores a la longitud objetivo"""
    secuencias_padded = []
    for seq in secuencias:
        if len(seq) < longitud_objetivo:
            # Añadir N's al final hasta alcanzar la longitud objetivo
            padding_necesario = longitud_objetivo - len(seq)
            seq_padded = seq + caracter_padding * padding_necesario
        else:
            # Si ya tiene 1600 o más, mantener como está
            seq_padded = seq[:longitud_objetivo]  # Truncar si es mayor a 1600
        secuencias_padded.append(seq_padded)
    return secuencias_padded

# Configuración para generar los nuevos conjuntos de datos
random.seed(42)  # Para reproducibilidad

print("Generando secuencias...")

# Grupo 1: 25,000 secuencias de 3200 caracteres con ACGNT
print("Generando 25,000 secuencias de 3200 caracteres con nucleótidos ACGNT...")
secuencias_3200 = generador_secuencias(
    n=25000, 
    l=3200, 
    nucleotidos="ACGNT"
)

# Crear DataFrame y guardar
df_3200 = pd.DataFrame({'secuencias': secuencias_3200})
df_3200.to_csv('secuencias_25k_3200bp_ACGNT.csv', index=False)
print(f"Guardado: secuencias_25k_3200bp_ACGNT.csv - {len(secuencias_3200)} secuencias")

# Grupo 2: 25,000 secuencias entre 1300-1600 caracteres con padding a 1600
print("Generando 25,000 secuencias de longitud variable (1300-1600) con nucleótidos ACGNT...")
secuencias_variables = generador_secuencias_variables(
    n=25000, 
    min_l=1300, 
    max_l=1600, 
    nucleotidos="ACGT"
)

# Aplicar padding a todas las secuencias que sean menores a 1600
print("Aplicando padding con 'N' hasta 1600 caracteres...")
secuencias_padded = aplicar_padding(secuencias_variables, longitud_objetivo=1600, caracter_padding='N')

# Crear DataFrame y guardar
df_padded = pd.DataFrame({'secuencias': secuencias_padded})
df_padded.to_csv('secuencias_25k_1300-1600bp_padded_ACGT.csv', index=False)
print(f"Guardado: secuencias_25k_1300-1600bp_padded_ACGT.csv - {len(secuencias_padded)} secuencias")

# Verificar estadísticas de las secuencias generadas
print("\n" + "="*50)
print("ESTADÍSTICAS DE LAS SECUENCIAS GENERADAS")
print("="*50)

print("\nGrupo 1 - Secuencias de 3200 bp:")
print(f"  Número de secuencias: {len(secuencias_3200)}")
print(f"  Longitud de todas las secuencias: {len(secuencias_3200[0])} bp")
print(f"  Nucleótidos utilizados: ACGNT")
print(f"  Ejemplo de secuencia: {secuencias_3200[0][:50]}...")

print("\nGrupo 2 - Secuencias variables con padding:")
print(f"  Número de secuencias: {len(secuencias_padded)}")
print(f"  Longitud después del padding: {len(secuencias_padded[0])} bp")
print(f"  Nucleótidos utilizados: ACGNT + N (padding)")

# Analizar cuántas secuencias necesitaron padding
longitudes_originales = [len(seq) for seq in secuencias_variables]  # Longitudes antes del padding
secuencias_con_padding = sum(1 for longitud in longitudes_originales if longitud < 1600)
print(f"  Secuencias que necesitaron padding: {secuencias_con_padding}")
print(f"  Secuencias que ya tenían 1600 bp: {25000 - secuencias_con_padding}")
print(f"  Longitud mínima original: {min(longitudes_originales)} bp")
print(f"  Longitud máxima original: {max(longitudes_originales)} bp")

# Ejemplo mostrando el padding aplicado
ejemplo_seq_original = secuencias_variables[0]
ejemplo_seq_padded = secuencias_padded[0]
print(f"  Ejemplo - Original ({len(ejemplo_seq_original)} bp): {ejemplo_seq_original[:30]}...")
print(f"  Ejemplo - Con padding ({len(ejemplo_seq_padded)} bp): {ejemplo_seq_padded[:30]}...{ejemplo_seq_padded[-10:]}")

# Opcional: Mostrar distribución de longitudes originales antes del padding
longitudes_unicas = {}
for longitud in longitudes_originales:
    longitudes_unicas[longitud] = longitudes_unicas.get(longitud, 0) + 1

print(f"\nDistribución de longitudes originales (primeras 10):")
for longitud in sorted(longitudes_unicas.keys())[:10]:
    print(f"  {longitud} bp: {longitudes_unicas[longitud]} secuencias")

print("\n¡Generación de secuencias completada exitosamente!")
```
