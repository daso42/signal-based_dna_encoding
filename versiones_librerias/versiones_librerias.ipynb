```python
# Python 3.10.15
"""
Script simple para obtener las versiones de las librerías principales del proyecto
"""

# Importar las librerías y obtener sus versiones
def get_versions():
    versions = {}
    
    try:
        import numpy as np
        versions['numpy'] = np.__version__
    except ImportError:
        versions['numpy'] = 'No instalada'
    
    try:
        import pandas as pd
        versions['pandas'] = pd.__version__
    except ImportError:
        versions['pandas'] = 'No instalada'
    
    try:
        import scipy
        versions['scipy'] = scipy.__version__
    except ImportError:
        versions['scipy'] = 'No instalada'
    
    try:
        import Bio
        versions['biopython'] = Bio.__version__
    except ImportError:
        versions['biopython'] = 'No instalada'
    
    try:
        import duckdb
        versions['duckdb'] = duckdb.__version__
    except ImportError:
        versions['duckdb'] = 'No instalada'
    
    try:
        import plotly
        versions['plotly'] = plotly.__version__
    except ImportError:
        versions['plotly'] = 'No instalada'
    
    try:
        import sklearn
        versions['scikit-learn'] = sklearn.__version__
    except ImportError:
        versions['scikit-learn'] = 'No instalada'
    
    try:
        import xgboost
        versions['xgboost'] = xgboost.__version__
    except ImportError:
        versions['xgboost'] = 'No instalada'
    
    try:
        import pywt
        versions['PyWavelets'] = pywt.__version__
    except ImportError:
        versions['PyWavelets'] = 'No instalada'
    
    try:
        import memory_profiler
        versions['memory-profiler'] = memory_profiler.__version__
    except ImportError:
        versions['memory-profiler'] = 'No instalada'
    
    try:
        import psutil
        versions['psutil'] = psutil.__version__
    except ImportError:
        versions['psutil'] = 'No instalada'
    
    return versions

if __name__ == "__main__":
    print("Versiones de librerías utilizadas en el proyecto:")
    print("=" * 50)
    
    versions = get_versions()
    
    for library, version in versions.items():
        print(f"{library:<20}: {version}")
    
    print("\nPython version:")
    import sys
    print(f"Python {sys.version}")
    
    # Generar requirements.txt básico
    print("\nGenerando requirements.txt...")
    with open("requirements.txt", "w") as f:
        f.write("# Requirements básicos para el proyecto\n")
        for library, version in versions.items():
            if version != 'No instalada':
                f.write(f"{library}=={version}\n")
    
    print("Archivo requirements.txt generado.")
```
