```python
dft=duckdb.sql("""
SELECT 
    modelo,
    encoding,
    ROUND(AVG(tiempo), 2) as avg_training_time,
    ROUND(STDDEV(tiempo), 2) as sd
FROM dft
GROUP BY modelo, encoding
""").to_df()
```

```python
duckdb.sql("""
select 
    f.modelo,
    d.encoding,
    f.avg_training_time,
    d.prediction_time,
    from df d
join dft f on f.encoding=d.encoding
where f.modelo = d.algorithm and f.encoding=d.encoding
""").to_df().to_csv('results/avg_training_eval_times.csv', index=False)
```
